```python
#import all of the perf text files here
import pandas as pd


file_path="/content/perfmem1.txt"
file_path2="/content/perfmem2.txt"
file_path3="/content/perfmem3.txt"
file_path4="/content/perfmem4.txt"
file_path5="/content/perfmem5.txt"
```

```python
def str_to_num(elements):
    for i in range(len(elements)):
        if ',' in elements[i]:
            elements[i] = float(elements[i].replace(",",""))
    return elements



with open(file_path, 'r') as f:
    lines = f.readlines()
first_time_stamp = float(lines[3].split()[0])
event_collected = set()
for line in lines[3:]:
    value = line.split()
    if float(value[0]) == first_time_stamp:
        event_collected.add(value[2])
    else:
        break

df_headers = ['time'] + list(event_collected)
data_stat= pd.DataFrame(columns=df_headers)
current_time_stamp = first_time_stamp
buffer = {'time':current_time_stamp}
for line in lines[3:]:
    if line[0] =='#' or len(line) == 1:
        continue
    elements = line.split()
    k=0
    k=len(elements)
    #print(elements[0],elements[1],elements[2])
    #if (k<=2):
    elements = str_to_num(elements[0:3])
    if elements[0] != current_time_stamp:
        data_stat.loc[len(data_stat)] = buffer
        current_time_stamp = elements[0]
        buffer['time'] = current_time_stamp
    buffer[elements[2]] = elements[1]

data_stat = data_stat.dropna(axis=0, how='any')
#data_stat
data_stat["time"] = pd.to_numeric(data_stat["time"], downcast="float")



with open(file_path2, 'r') as f:
    lines = f.readlines()
first_time_stamp = float(lines[3].split()[0])
event_collected = set()
for line in lines[3:]:
    value = line.split()
    if float(value[0]) == first_time_stamp:
        event_collected.add(value[2])
    else:
        break

df_headers = ['time'] + list(event_collected)
data_stat2 = pd.DataFrame(columns=df_headers)
current_time_stamp = first_time_stamp
buffer = {'time':current_time_stamp}
for line in lines[3:]:
    if line[0] =='#' or len(line) == 1:
        continue
    elements = line.split()
    k=0
    k=len(elements)
    #print(elements[0],elements[1],elements[2])
    #if (k<=2):
    elements = str_to_num(elements[0:3])
    if elements[0] != current_time_stamp:
        data_stat2.loc[len(data_stat2)] = buffer
        current_time_stamp = elements[0]
        buffer['time'] = current_time_stamp
    buffer[elements[2]] = elements[1]

data_stat2 = data_stat2.dropna(axis=0, how='any')
data_stat2["time"] = pd.to_numeric(data_stat2["time"], downcast="float")


with open(file_path3, 'r') as f:
    lines = f.readlines()
first_time_stamp = float(lines[3].split()[0])
event_collected = set()
for line in lines[3:]:
    value = line.split()
    if float(value[0]) == first_time_stamp:
        event_collected.add(value[2])
    else:
        break

df_headers = ['time'] + list(event_collected)
data_stat3 = pd.DataFrame(columns=df_headers)
current_time_stamp = first_time_stamp
buffer = {'time':current_time_stamp}
for line in lines[3:]:
    if line[0] =='#' or len(line) == 1:
        continue
    elements = line.split()
    k=0
    k=len(elements)
    #print(elements[0],elements[1],elements[2])
    #if (k<=2):
    elements = str_to_num(elements[0:3])
    if elements[0] != current_time_stamp:
        data_stat3.loc[len(data_stat3)] = buffer
        current_time_stamp = elements[0]
        buffer['time'] = current_time_stamp
    buffer[elements[2]] = elements[1]

data_stat3 = data_stat3.dropna(axis=0, how='any')
data_stat3["time"] = pd.to_numeric(data_stat3["time"], downcast="float")
data_stat3


with open(file_path4, 'r') as f:
    lines = f.readlines()
first_time_stamp = float(lines[3].split()[0])
event_collected = set()
for line in lines[3:]:
    value = line.split()
    if float(value[0]) == first_time_stamp:
        event_collected.add(value[2])
    else:
        break

df_headers = ['time'] + list(event_collected)
data_stat4 = pd.DataFrame(columns=df_headers)
current_time_stamp = first_time_stamp
buffer = {'time':current_time_stamp}
for line in lines[3:]:
    if line[0] =='#' or len(line) == 1:
        continue
    elements = line.split()
    k=0
    k=len(elements)
    #print(elements[0],elements[1],elements[2])
    #if (k<=2):
    elements = str_to_num(elements[0:3])
    if elements[0] != current_time_stamp:
        data_stat4.loc[len(data_stat4)] = buffer
        current_time_stamp = elements[0]
        buffer['time'] = current_time_stamp
    buffer[elements[2]] = elements[1]

data_stat4 = data_stat4.dropna(axis=0, how='any')
data_stat4["time"] = pd.to_numeric(data_stat4["time"], downcast="float")
data_stat4


with open(file_path5, 'r') as f:
    lines = f.readlines()
first_time_stamp = float(lines[3].split()[0])
event_collected = set()
for line in lines[3:]:
    value = line.split()
    if float(value[0]) == first_time_stamp:
        event_collected.add(value[2])
    else:
        break

df_headers = ['time'] + list(event_collected)
data_stat5 = pd.DataFrame(columns=df_headers)
current_time_stamp = first_time_stamp
buffer = {'time':current_time_stamp}
for line in lines[3:]:
    if line[0] =='#' or len(line) == 1:
        continue
    elements = line.split()
    k=0
    k=len(elements)
    #print(elements[0],elements[1],elements[2])
    #if (k<=2):
    elements = str_to_num(elements[0:3])
    if elements[0] != current_time_stamp:
        data_stat5.loc[len(data_stat5)] = buffer
        current_time_stamp = elements[0]
        buffer['time'] = current_time_stamp
    buffer[elements[2]] = elements[1]

data_stat5 = data_stat5.dropna(axis=0, how='any')
data_stat5["time"] = pd.to_numeric(data_stat5["time"], downcast="float")

df=pd.merge(data_stat,data_stat2, how='outer' ,left_index = True, right_index = True , suffixes = ('', '_right'))
df = df.drop(['time_right'], axis=1)
df=pd.merge(df,data_stat3, how='outer' ,left_index = True, right_index = True , suffixes = ('', '_right'))
df = df.drop(['time_right'], axis=1)
df=pd.merge(df,data_stat4, how='outer' ,left_index = True, right_index = True , suffixes = ('', '_right'))
df = df.drop(['time_right'], axis=1)
df=pd.merge(df,data_stat5, how='outer' ,left_index = True, right_index = True , suffixes = ('', '_right'))
df = df.drop(['time_right'], axis=1)
df.dropna(how='any', inplace=True)
df
```
